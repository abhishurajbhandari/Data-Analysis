```python
# Chatgpt

# 1. What is the average temperature over time?

df['Date/Time'] = pd.to_datetime(df['Date/Time']) # Converting column into datatime format
```

```python
avg_temp = df.groupby(df['Date/Time'].dt.date)['Temp_C'].mean() # We are grouping by datetime
                                                                # The "dt" will get the date time properties
                                                                # .date extracts just date excluding time
                                                                # We are taking mean for each group (date)
print(avg_temp)
```

```python
plt.figure(figsize= (18,6))
plt.plot(avg_temp.index, avg_temp.values, color = 'blue', marker = 'o')
plt.title('Average Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

```python
# 2. How does the temperature vary between different weather conditions?

temp_by_weather = df.groupby('Weather Condition')['Temp_C'].mean() # first groups by weather and then selects temp column before appying mean
                                                                    # By selecting temp first we are narrowing down the column first and then using mean
print(temp_by_weather)
```

```python
plt.figure(figsize=(18,6))
temp_by_weather.plot(kind = 'bar', color='skyblue')
plt.title('Average Temperature by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Are there seasonal patterns in temperature?

df['Month'] = df['Date/Time'].dt.month # Extract just month
```

```python
avg_monthly_temp = df.groupby('Month')['Temp_C'].mean()
print(avg_monthly_temp)
```

```python
plt.figure(figsize=(15,6))
plt.plot(avg_monthly_temp.index, avg_monthly_temp.values, color='green', marker='o') # plt plot creates a line plot
plt.title('Average Temperature by month')                                             # index represents the month
plt.xlabel('Month')                                                                   # values is the temp values
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.xticks(avg_monthly_temp.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
# 4. Is there a correlation between Dew Point Temperature and Relative Humidity?

correlation = df['Dew Point Temp_C'].corr(df['Temp_C'])              # first take x value and then corr to y values
print(f"Correlation between Dew Point and Humidity: {correlation}")  # yes there is a high positive correlation
                                                                    # If Dew Point increases Humidity increases as well
```
