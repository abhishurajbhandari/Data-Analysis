```python
# 16. average duration of movies
movies = df[df['type'] == 'Movie']   # only taking movies
print(movies)
```

```python
# Dropping nan values before processing
movies = movies.dropna(subset=['duration'])
```

```python
# Convert the 'duration' column for movies into numeric values
movies['duration_min'] = movies['duration'].str.replace(' min', '')   # replacing sting min with ''
```

```python
# Converting it into a int
movies['duration_min'] = movies['duration_min'].astype(int)
```

```python
avg_movie_duration = movies['duration_min'].mean()
print(avg_movie_duration)
```

```python
# 17. How does the duration of content vary by rating or genre?

avg_movie_duration_rating = movies.groupby('rating')['duration_min'].mean()
avg_movie_duration_rating
```

```python
# a. Without splitting the genre string
avg_movie_duration_genre = movies.groupby('listed_in')['duration_min'].mean()
avg_movie_duration_genre
```

```python
# 21. Is there any correlation between the release year and rating of a show?

# a. factorize the rating for individual numbers
df['rating_numeric'] = df['rating'].factorize()[0]   # 0 means the starting number factorize would put out
```
